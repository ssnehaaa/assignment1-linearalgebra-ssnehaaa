### Q2 Part (iii) — NumPy Function

Write a NumPy function that:
- Takes **any 2×2 matrix** and **any 2D vector** as inputs
- Returns the **transformed vector**
- Prints a **one-line description** of how the x and y components individually change

Test your function on matrix `A` and all four vectors above.

````python
# Input — do not modify
A = np.array([[2, 0], [0, 3]])
v1 = np.array([1, 0])
v2 = np.array([0, 1])
v3 = np.array([1, 1])
v4 = np.array([-2, 4])
````

````python
import numpy as np

def transform_vector(matrix, vector):
    # Ensure inputs are numpy arrays
    M = np.array(matrix)
    v = np.array(vector)

    # Perform the transformation
    transformed = M @ v

    # Print the individual component changes
    print(f"x changed from {v[0]} to {transformed[0]} (scaled by {matrix[0][0] if matrix[0][1]==0 else 'dynamic'}), "
          f"y changed from {v[1]} to {transformed[1]} (scaled by {matrix[1][1] if matrix[1][0]==0 else 'dynamic'}).")

    return transformed

# Example Usage:
A = [[2, 0], [0, 3]]
v3 = [1, 1]
transform_vector(A, v3)
````

## Q3 — Tiny Recommendation System

Each user's taste is represented as a vector of 4 scores across 4 genres — **action, romance, thriller, comedy** — all between 0 and 1.

```
User A: [0.9, 0.1, 0.8, 0.2]
User B: [0.1, 0.9, 0.2, 0.8]
User C: [0.8, 0.2, 0.9, 0.1]
User D: [0.5, 0.5, 0.5, 0.5]
```

### Q3 Part (i) — Similarity Matrix

Compute the **full similarity matrix** — every user's similarity score against every other user — using only a **single NumPy operation** (no loops).

What linear algebra concept makes this possible in one line?

````python
# Input — do not modify
users = np.array([
    [0.9, 0.1, 0.8, 0.2],  # User A
    [0.1, 0.9, 0.2, 0.8],  # User B
    [0.8, 0.2, 0.9, 0.1],  # User C
    [0.5, 0.5, 0.5, 0.5],  # User D
])
````

````python
import numpy as np

# Define user taste matrix
users = np.array([
    [0.9, 0.1, 0.8, 0.2],  # User A
    [0.1, 0.9, 0.2, 0.8],  # User B
    [0.8, 0.2, 0.9, 0.1],  # User C
    [0.5, 0.5, 0.5, 0.5]   # User D
])

# Single NumPy operation for the full similarity matrix
similarity_matrix = users @ users.T
print(similarity_matrix)
````

**What linear algebra concept makes this possible in one line?**


The linear algebra concept that makes this possible in a single line is the matrix-matrix multiplication (or the outer dot product of a matrix with its transpose).

By multiplying the user matrix by its own transpose, we compute the dot product of every row (user) against every column (which is also every user).

### Q3 Part (iii) — The Problem with Raw Dot Products

A senior engineer tells you:
> *"Raw dot products are a bad measure of similarity — you need to normalize."\*

1. What is the **problem** with raw dot products?
2. Construct a **specific example using NumPy** that demonstrates the flaw.
3. Fix it.

> **Hint:** Take a user with scores `[1.8, 0.2, 1.6, 0.4]`

*Write your answer here (double-click to edit this cell)*

````python
user_A = np.array([0.9, 0.1, 0.8, 0.2])
user_C = np.array([0.8, 0.2, 0.9, 0.1])
user_A_intense = np.array([1.8, 0.2, 1.6, 0.4])

# Even though A and C have almost identical tastes, look at raw dot products:
print("Dot product (A, C):", np.dot(user_A, user_C))          # Result: 1.48
print("Dot product (A_intense, C):", np.dot(user_A_intense, user_C))  # Result: 2.96
````

1. The Problem: Raw dot products are sensitive to the magnitude (scale) of the vectors, not just the preference profile. If a user likes the exact same genres but happens to rate things more intensely (or consumes a lot more content), their vector magnitude increases, artificially inflating their similarity score.
2. Demonstration of the Flaw:
Let's introduce "User A-Intense", who has the exact same taste profile as User A but with doubled scores: [1.8, 0.2, 1.6, 0.4].
Flaw: User A-Intense appears "twice as similar" to User C than User A is, simply because their rating scales are larger.
3. Flaw: User A-Intense appears "twice as similar" to User C than User A is, simply because their rating scales are larger.The Fix (Cosine Similarity):We fix this by normalizing the vectors to unit length before taking the dot product, computing Cosine Similarity:
{Cosine Similarity} = (u.v)/(||u||x||v||)

````python
# Function to normalize matrix rows to unit vectors
norms = np.linalg.norm(users, axis=1, keepdims=True)
normalized_users = users / norms

# Compute normalized similarity matrix
cosine_similarity_matrix = normalized_users @ normalized_users.T
print(cosine_similarity_matrix)
````
